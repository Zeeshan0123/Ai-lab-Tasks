```python

# This Program is intelligent its also save the previous States

sprinkle = False
Alarm = False
T_Alarm = False
temperature = False
program = True

#Normal temperature range is 30-60
#High temperature range is greater then 60
#lower temperature is less then 30


def smoke_dec(smok):
    global Alarm,sprinkle
    if(smok == 'Y'):
            print("Smoke is detected and Alarm is ON")
            Alarm = True
            print("Sprinkler system will activate")
            sprinkle = True
            
    elif(smok == 'N'):

            if(Alarm == True & sprinkle == True):
                Alarm = False
                sprinkle = False
                print("No Smoke is Detected Now Alarm and and Sprinkler is Turned OFF")
            else:
                print("No Smoke is detected")
                
def tem_det(temp):
        global T_Alarm,temperature                                                               
        temp = int(temp)
        if(temp in range(30,60)):
            print("Temperature is Normal")
            if(T_Alarm == True & temperature == True):
                print("Alarm will be turned OFF ")
                T_Alarm = False
                temperature = False
        elif(temp > 60):
            print("Temperature is High Alarm will be sound")
            T_Alarm = True
            temperature = True
        else:
            print("Temperature is Low")
        
    

while program == True:
    
    temp = input("Enter the temperature of the Single floor:")
    print(f"The temperature for the Office buiding floor is {temp}")

    smok = input("Is there any Smoke: Y/N\n").upper()

    
    tem_det(temp)
    smoke_dec(smok)  #Calling Both Functions
    
    
    final = input("Do you want to Continue Y/N\n").upper()
    if(final == 'Y'):
        program = True
    else:
        program = False

print("Thanks You :)")
        

```
